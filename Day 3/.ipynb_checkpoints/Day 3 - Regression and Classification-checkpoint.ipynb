## Calculate MSE

```python
import seaborn as sn
import numpy as np

tipsdata=sn.load_dataset('tips')
```

```python
totalBill = tipsdata.total_bill
tips = tipsdata.tip
```

```python
npX = np.array(list(totalBill))
npY = np.array(list(tips))
# print(npData)
xMean, yMean = npX.mean(axis=0), npY.mean(axis=0)
xMean, yMean 
```

```python
theta1 = ((npX - xMean) * (npY - yMean)).sum() / ((npX - xMean) ** 2).sum()

theta0 = yMean - theta1 * xMean
theta0, theta1
```

```python
npYPred = theta0 + theta1 * npX

MSE = ((npY - npYPred)**2).sum() / npY.size
MSE
```

```python
import matplotlib.pyplot as plt

plt.scatter(npX, npY)
plt.plot(npX, npYPred, color='red')
plt.show()
```
